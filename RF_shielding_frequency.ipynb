```python
def error_propagation(u,v):
    
    # x = u/v
    
    x_std = np.sqrt((u.std()/v.mean())**2+(u.mean()*v.std()/(v.mean()**2))**2-2*u.mean()*np.cov(u,v)[0, 1]/(v.mean()**3))
    
    x_mean = u.mean()/v.mean()
    
    return x_mean, x_std
```
